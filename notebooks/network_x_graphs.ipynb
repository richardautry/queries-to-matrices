```python
import networkx as nx
G = nx.Graph()
```

```python
G.add_nodes_from([
    ("CargoShipA", {"type": "RED", "order": 0, "dock_time": 900}),
    ("CargoShipB", {"type": "BLUE", "order": 1, "dock_time": 1100}),
])
```

```python
G.add_edge("CargoShipA", "CargoShipB")
```

```python
for node in G.nodes:
    print(G.nodes[node])
```

```python
import matplotlib.pyplot as plt
```

```python
# subax1 = plt.subplot(121)
def build_pos(graph):
    pos_dict = {}
    for node in list(graph.nodes):
        pos_dict[node] =  [graph.nodes[node]["dock_time"], graph.nodes[node]["order"]]
    return pos_dict


fig, ax = plt.subplots()
plt.xlabel("Dock Time")
plt.ylabel("Order")
plt.title("Cargo Ship Conflict Graph")
nx.draw(G, with_labels=True, font_weight='bold', pos=build_pos(G), ax=ax)
limits = plt.axis('on')
ax.tick_params(left=True, bottom=True, labelleft=True, labelbottom=True)
ax.invert_yaxis()
ax.set_xlim(500, 1400)
plt.savefig("cargo_ship_simple_conflict_graph.png")
```

```python
G.add_nodes_from([
    ("CargoShipC", {"type": "RED", "order": 2, "dock_time": 1500}),
    ("CargoShipD", {"type": "BLUE", "order": 3, "dock_time": 1600})
])

G.add_edge("CargoShipB", "CargoShipC")
G.add_edge("CargoShipA", "CargoShipD")

```

```python
fig, ax = plt.subplots()
plt.xlabel("Dock Time")
plt.ylabel("Order")
plt.title("Cargo Ship Conflict Graph")
nx.draw(G, with_labels=True, font_weight='bold', pos=build_pos(G), ax=ax)
limits = plt.axis('on')
ax.tick_params(left=True, bottom=True, labelleft=True, labelbottom=True)
ax.invert_yaxis()
ax.set_xlim(500, 2000)
plt.savefig("cargo_ship_complex_conflict_graph.png")
```
